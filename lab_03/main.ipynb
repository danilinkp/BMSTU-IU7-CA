# Приближенное аналитическое решение обыкновенного дифференциального уравнения

```python
x = np.linspace(0, 1, 100)

# Для n = 1
coeffs_n1 = solve_ode_least_squares(2)
y_n1 = evaluate_solution(x, 2, coeffs_n1)
print("Коэффициенты для n=2:", *coeffs_n1)

# Для n = 2
coeffs_n2 = solve_ode_least_squares(3)
y_n2 = evaluate_solution(x, 3, coeffs_n2)
print("Коэффициенты для n=3:", *coeffs_n2)

plt.plot(x, y_n1, label='m=2')
plt.plot(x, y_n2, label='m=3')
plt.legend()
plt.show()
```
